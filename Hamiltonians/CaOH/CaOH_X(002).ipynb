```julia
using QuantumStates, UnitsToValue
```

```julia
H_operator = :(
    TX * Identity +
    BX * Rotation + 
    DX * RotationDistortion + 
    Î³X * SpinRotation +
    bFX * Hyperfine_IS +
    cX * (Hyperfine_Dipolar/3) +
    s * basis_splitting
);
```

```julia
parameters = @params begin
    TX = 2 * 3847 * c * 1e2 # we'll just take the energy to be 2 Ã— 3847 cmâ»Â¹
    BX = 0.333458 * c * 1e2
    DX = 0.3930e-6 * c * 1e2
    Î³X = 0.001184 * c * 1e2
    bFX = 2.602 * 1e6
    cX = 2.053 * 1e6
    s = 1e-9 * c * 1e2
end;
```

```julia
QN_bounds = (
    v_1 = 0,
    v_2 = 0,
    v_3 = 3,
    S = 1/2,
    I = 1/2,
    Î› = 0,
    â„“ = 0,
    K = 0,
    N = 0:5
)
basis = enumerate_states(HundsCaseB_LinearMolecule, QN_bounds)
;
```

```julia
CaOH_X002 = @time Hamiltonian(basis=basis, operator=H_operator, parameters=parameters)
@time full_evaluate!(CaOH_X002)
@time solve!(CaOH_X002)
;
```

```julia
@time save_to_file(CaOH_X002, "CaOH_X002", "")
```
